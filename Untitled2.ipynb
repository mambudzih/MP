```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
from google.colab import drive
```

```python
# Mount Google Drive to access the model and images
drive.mount('/content/drive')
```

```python
# Load the pre-trained model
model_path = '/content/drive/MyDrive/saved models/traine_model.keras'
model = tf.keras.models.load_model(model_path)
```

```python
# Load and display the test image
image_path = "/content/drive/MyDrive/Colab Notebooks/test/test/gls3.jpg"
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
plt.imshow(img)
plt.title("Test Image")
plt.axis('off')  # Hide axes
plt.show()
```

```python
# Preprocess the image for prediction
image = tf.keras.preprocessing.image.load_img(image_path, target_size=(128, 128))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr])  # Add batch dimension
print(f"Input shape: {input_arr.shape}")
```

```python
# Make prediction
prediction = model.predict(input_arr)
print(f"Prediction shape: {prediction.shape}")
```

```python
# Get the class index with the highest probability
result_index = np.argmax(prediction)
```

```python
# Define class names
class_names = [
    'Corn_(maize)___Cercospora_leaf_spot Gray_leaf_spot',
    'Corn_(maize)___Common_rust_',
    'Corn_(maize)___Northern_Leaf_Blight',
    'Corn_(maize)___healthy'
]
```

```python
# Get the predicted disease name
model_prediction = class_names[result_index]
```

```python
# Define treatments based on disease predictions
def get_treatment(disease_class):
    if disease_class == 0:
        return "Treatment for Cercospora leaf spot: Apply fungicide and remove infected leaves."
    elif disease_class == 1:
        return "Treatment for Common rust: Use resistant varieties and apply fungicides."
    elif disease_class == 2:
        return "Treatment for Northern Leaf Blight: Implement crop rotation and apply fungicides."
    elif disease_class == 3:
        return "No treatment needed (Healthy): Maintain good agricultural practices."
    else:
        return "No treatment available"
```

```python
# Get the treatment recommendation
treatment_recommendation = get_treatment(result_index)
```

```python
# Display the image with the predicted disease name
plt.imshow(img)
plt.title(f"Disease Name: {model_prediction}", fontsize=14)
plt.axis('off')  # Hide axes
```

```python
# Add the treatment recommendation below the image
plt.figtext(0.5, -0.1, f"Recommended Treatment: {treatment_recommendation}", wrap=True, horizontalalignment='center', fontsize=12)
```

```python
# Show the plot
plt.show()
```

```python
# Print the predicted disease name and treatment recommendation
print(f"Predicted Disease: {model_prediction}")
print(f"Recommended Treatment: {treatment_recommendation}")
```
